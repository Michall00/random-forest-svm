# Wine Norm

```python
wine_df['quality'].value_counts().plot(kind='bar')
```

```python
wine_df.loc[wine_df['quality'] > 6].count()
```

```python
wine_df["quality"] = wine_df["quality"].apply(lambda x: 1 if x > 6 else 0)
```

```python
wine_df["quality"]
```

```python
wine_df.hist()
```

```python
log1_norm_columns = ["residual_sugar", "chlorides", "free_sulfur_dioxide", "total_sulfur_dioxide"]  # Rozkad prawoskony
standarize_columns = ["fixed_acidity", "volatile_acidity", "citric_acid", "density", "pH", "sulphates", "alcohol"] # Rozkad przypominajcy normalny
```

```python
for column in log1_norm_columns:
    wine_df[column] = wine_df[column].apply(lambda x: np.log1p(x))

scaler = StandardScaler()
wine_df[standarize_columns] = scaler.fit_transform(wine_df[standarize_columns])
```

```python
plt.figure(figsize=(15, 15))
wine_df.hist()
```

# Iris Norm

```python
iris_df["class"].value_counts().plot(kind='bar')
```

```python
iris_df["class"] = iris_df["class"].apply(lambda x: 1 if x == "Iris-setosa" else 0)
```

```python
scaler = StandardScaler()
iris_df[iris_df.columns[:-1]] = scaler.fit_transform(iris_df[iris_df.columns[:-1]])
```

# Churn norm

```python
churn_df.hist()
```

```python
columns_to_log = [
    'Call  Failure', 'Seconds of Use', 'Frequency of use',
    'Frequency of SMS', 'Customer Value', 'Charge  Amount', 'Distinct Called Numbers',

]
columns_to_standardize = ['Subscription  Length', 'Age',]
```

```python
churn_df.nunique()
```

```python
for column in columns_to_log:
    churn_df[column] = churn_df[column].apply(lambda x: np.log1p(x))
```

```python
scaler = StandardScaler()
churn_df[columns_to_standardize] = scaler.fit_transform(churn_df[columns_to_standardize])
churn_df.hist()
```
